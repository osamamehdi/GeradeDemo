```python
import matplotlib.pyplot as plt

xs = []
ys_1 = []
ys_2 = []

m_1 = 1
m_2 = 1

t_1 = 1
t_2 = 2

# Berechnung der y-Werte der darzustellenden Geraden
for x in range(0, 3):
    xs.append(x) # Wertebereich der x-Achse festlegen
    ys_1.append((m_1 * x) + t_1)
    ys_2.append((m_2 * x) + t_2)
    
plt.plot(xs, ys_1, "r-") # Rote Gerade
plt.plot(xs, ys_2, "b-") # Blaue Gerade
plt.grid(True)
plt.show()
```
